## Step 1: Imports

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Ethan-Mitchell-Data-Analyst/Python-Data-Cleaning-Project/main/data/Messy_Employee_dataset.csv"
df = pd.read_csv(url)

print(df.head())
print(df.info())
```

Business Scenario: Provide a well-formatted list of employees [...] for others to group data as they need (e.g. combinations of Department and Region).

## Step 3: Data Cleaning

- [x] Drop duplicates (drop rows we don't need)
- [ ] Drop columns we don't need (for now I want to keep all the columns)
- [x] Strip non-alphabet characters/non-numbers for each column as relevant
- [x] Split Department_Region column into Department and Region
- [x] Convert Join_Date strings to datetime format
- [x] Change Salary column to include US$ in the column name
- [x] Format Phone column for making international calls
- [x] Check unique values in Status, Performance_Score and Remote_Work columns - there shouldn't be any typos
- [ ] Re-index Department and Region columns
- [ ] Remove Department_Region compound column, now useless

```python
#Drop duplicates
df = df.drop_duplicates()
```

```python
#Strip non-alphabet characters/non-numbers for each column as relevant
df["Employee_ID"] = df["Employee_ID"].str.replace('^a-zA-Z0-9', '')
df["First_Name"] = df["First_Name"].str.replace('^a-zA-Z', '')
df["Last_Name"] = df["Last_Name"].str.replace('^a-zA-Z', '')
df["Department_Region"] = df["Department_Region"].str.replace('^a-zA-Z-', '')
df["Status"] = df["Status"].str.replace('^a-zA-Z', '')
df["Join_Date"] = df["Join_Date"].str.replace('^0-9/', '')
df["Performance_Score"] = df["Performance_Score"].str.replace('^a-zA-Z', '')
```

```python
#Split Department_Region column into Department and Region
df[["Department", "Region"]] = df["Department_Region"].str.split('-', expand=True)
#df
```

```python
#Convert Join_Date strings to datetime format
df["Join_Date"] = pd.to_datetime(df["Join_Date"])
df["Join_Date"]
```

```python
#Change Salary column to include US$ in the column name
df.rename(columns={'Salary': 'Salary_US$'})
```

```python
#Format Phone column for making international calls, i.e. +1 (xxx) xxx-xxxx, this requires string parsing
df["Phone"] = df["Phone"].map(str).str.strip("-")
df["Phone"].str.replace('^0-9', '')
```

```python
df["Phone"] = df["Phone"].apply(lambda x: "+1 (" + x[0:3] + ") " + x[3:6] + "-" + x[6:10])
df["Phone"]
```

```python
#Check unique values in Status, Performance_Score and Remote_Work columns - there shouldn't be any typos
print(df["Status"].unique())
print(df["Performance_Score"].unique())
print(df["Remote_Work"].unique())
```
